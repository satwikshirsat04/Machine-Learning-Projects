PRE PROCESS THE DATA

```python
Adj_Close= google_data["Close"]
max(Adj_Close.values),min(Adj_Close.values)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(Adj_Close)
len(scaled_data)
```

```python
x_data = []
y_data = []

for i in range(100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

x_data,y_data = np.array(x_data), np.array(y_data)
```

```python
print(x_data[0],y_data[0]) #100 Features and 1 Output
```

```python
len(x_data),len(y_data)  # -100 rows (5031 - 100)
```

TRAIN TEST SPLIT

```python
spliting_length = int(len(x_data)*0.70)
```

```python
spliting_length
```

```python
x_train = x_data[:spliting_length]
y_train = y_data[:spliting_length]
```

```python
len(x_train)
```

```python
x_test = x_data[spliting_length:]
y_test = y_data[spliting_length:]
```

```python
len(x_test)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

TRAIN THE MODEL USING KERAS

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(128,return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=20)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
inverse_pred = scaler.inverse_transform(predictions)
inverse_pred
```

```python
inverse_y_test = scaler.inverse_transform(y_test)
inverse_y_test
```

```python
len(inverse_y_test)
```

ERROR RATE WITH MEAN SQUARE ERROR

```python
mse = np.sqrt(np.mean((inverse_pred - inverse_y_test)**2))
mse
```

PLOTTING GRAPH BETWEEN ORIGINAL AND PREDICTED VALUES (TESTING DATASET)

```python
ploting_data = pd.DataFrame(
    {
        "Original_Test_Data" : inverse_y_test.reshape(-1),
        "Predicted_Test_Data" : inverse_pred.reshape(-1)
    },
    index = google_data.index[spliting_length+100:]
)
ploting_data.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(ploting_data['Original_Test_Data'],label="Original Test_Data",color="r")
plt.plot(ploting_data['Predicted_Test_Data'],label="Predicted Test_Data",color="g")
plt.xlabel("years")
plt.ylabel("Close")
plt.grid=True
plt.legend()
plt.title("Original v/s Predictions Test Data (2.46% Error Rate)")
plt.savefig("Original vs Predictions(Test Data)",dpi=300)
plt.show()
```

```python
plot_graph(pd.concat([Adj_Close[:spliting_length+100],ploting_data],axis=0),"Full Data")
```
