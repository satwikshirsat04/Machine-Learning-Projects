```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
google_data = pd.read_csv("google_stock_data.csv")
```

```python
plt.figure(figsize=(15,4),)
plt.plot(google_data["Close"],label="Close Price")
plt.xlabel("years")
plt.ylabel("Close")
plt.grid=True
plt.legend()
plt.title("Close Price of GOOGLE Stock")
plt.show()
```

```python
def plot_graph(value,column):
    plt.figure(figsize=(15,4))
    plt.plot(value)
    plt.xlabel("years")
    plt.ylabel("column")
    plt.legend()
    plt.title(f"{column} Price of GOOGLE Stock")
    plt.show()
```

```python
for col in google_data.columns:
    plot_graph(google_data[col],col)
```
